# **3. Consistencia en valores categóricos: Identifica y corrige cualquier inconsistencia en los valores categóricos (por ejemplo, ‘Guard’, ‘guard’, ‘GUARD’).**

```python
# Paso 1: Importo la librería para conectarme a Google Drive
from google.colab import drive

# Me conecto a mi Google Drive
drive.mount('/content/drive')

# Paso 2: Importo pandas
import pandas as pd

# Enlaces de los archivos actualizados
# Temporada regular
url_regular = 'https://drive.google.com/uc?id=10Z1M-L_JQgeYSsFB2V22bq3hBroYwHvw'
# Playoffs
url_playoff = 'https://drive.google.com/uc?id=10XjB66anvilzdwb0bgtD33u7mHZI1xUP'

# Paso 3: Uso pandas para leer los archivos CSV desde los enlaces de Google Drive
# Agrego encoding para evitar problemas con caracteres raros
df_regular = pd.read_csv(url_regular, encoding='ISO-8859-1', sep=';')
df_playoff = pd.read_csv(url_playoff, encoding='ISO-8859-1', sep=';')

# Paso 4: Consistencia en valores categóricos
# Normalizar valores de la columna 'Pos'
df_regular['Pos'] = df_regular['Pos'].str.strip().str.capitalize()  # Quitar espacios y capitalizar
df_playoff['Pos'] = df_playoff['Pos'].str.strip().str.capitalize()  # Quitar espacios y capitalizar

# Confirmar consistencia de valores en 'Pos' antes de la corrección
print(f"Valores únicos en la columna 'Pos' - Temporada Regular antes de la corrección:\n{df_regular['Pos'].unique()}")
print(f"Valores únicos en la columna 'Pos' - Playoffs antes de la corrección:\n{df_playoff['Pos'].unique()}")

# Reemplazar inconsistencias específicas en 'Pos'
inconsistencias_guard = ['Guard', 'Guard ', 'GUARD', 'guard']
df_regular['Pos'] = df_regular['Pos'].replace(inconsistencias_guard, 'Guard')  # Corrige inconsistencias
df_playoff['Pos'] = df_playoff['Pos'].replace(inconsistencias_guard, 'Guard')  # Corrige inconsistencias

# Confirmar la corrección
print(f"Valores únicos en la columna 'Pos' - Temporada Regular después de la corrección:\n{df_regular['Pos'].unique()}")
print(f"Valores únicos en la columna 'Pos' - Playoffs después de la corrección:\n{df_playoff['Pos'].unique()}")

# Paso 5: Normalización de la columna 'Tm'
# Convertir todos los valores a mayúsculas para asegurar consistencia
df_regular['Tm'] = df_regular['Tm'].str.upper().str.strip()
df_playoff['Tm'] = df_playoff['Tm'].str.upper().str.strip()

# Confirmar valores únicos en 'Tm'
print(f"Valores únicos en la columna 'Tm' - Temporada Regular:\n{df_regular['Tm'].unique()}")
print(f"Valores únicos en la columna 'Tm' - Playoffs:\n{df_playoff['Tm'].unique()}")

# Paso 6: Validación de rango para columnas numéricas
# Asegurarse de que 'FG%', 'PTS' estén dentro de los límites esperados
df_regular = df_regular[(df_regular['FG%'] >= 0) & (df_regular['FG%'] <= 1)]
df_playoff = df_playoff[(df_playoff['FG%'] >= 0) & (df_playoff['FG%'] <= 1)]

df_regular = df_regular[(df_regular['PTS'] >= 0)]
df_playoff = df_playoff[(df_playoff['PTS'] >= 0)]

# Confirmar el número de filas después de la validación
print(f"Filas después de la validación - Temporada Regular: {df_regular.shape[0]}")
print(f"Filas después de la validación - Playoffs: {df_playoff.shape[0]}")

# Paso 7: Exportamos a Google Drive el nuevo dataset
ruta_regular = '/content/drive/My Drive/Colab Notebooks/Examen 01/df_regular_consistencia.csv'
ruta_playoff = '/content/drive/My Drive/Colab Notebooks/Examen 01/df_playoff_consistencia.csv'

# Exportar el dataset con consistencia a CSV
df_regular.to_csv(ruta_regular, sep=';', index=False)  # Asegúrate de incluir el separador
df_playoff.to_csv(ruta_playoff, sep=';', index=False)  # Asegúrate de incluir el separador

# Confirmación de exportación
print(f"Dataset de la Temporada Regular exportado a: {ruta_regular}")
print(f"Dataset de los Playoffs exportado a: {ruta_playoff}")

### CITAS ###
# 1. Pandas Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
# 2. Google Colab - Connecting to Google Drive: https://colab.research.google.com/notebooks/io.ipynb
# 3. Pandas read_csv documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
# 4. Pandas Documentation on DataFrame.to_csv: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html
```

# **4. Manejo de valores faltantes: Identifica y maneja cualquier valor faltante en el dataset. Rellena los valores faltantes con un marcador adecuado para el tipo de dato.**

```python
# Paso 1: Importo la librería para conectarme a Google Drive
from google.colab import drive

# Me conecto a mi Google Drive
drive.mount('/content/drive')

# Paso 2: Importo pandas
import pandas as pd

# Enlaces de los archivos actualizados
# Temporada regular
url_regular = 'https://drive.google.com/uc?id=10hQY1pIaJQX6x2zGhjAH8awsbvCL1N2B'
# Playoffs
url_playoff = 'https://drive.google.com/uc?id=10aG1zmtp-DB8c4HLlfYy9Y-0sSatihiK'

# Paso 3: Uso pandas para leer los archivos CSV desde los enlaces de Google Drive
# Agrego encoding para evitar problemas con caracteres raros
df_regular = pd.read_csv(url_regular, encoding='ISO-8859-1', sep=';')
df_playoff = pd.read_csv(url_playoff, encoding='ISO-8859-1', sep=';')

# Paso 4: Identificar valores faltantes
missing_values_regular = df_regular.isnull().sum()
missing_values_playoff = df_playoff.isnull().sum()

# Crear un DataFrame para los resultados de los valores faltantes
faltantes_regular = missing_values_regular[missing_values_regular > 0].reset_index()
faltantes_playoff = missing_values_playoff[missing_values_playoff > 0].reset_index()

# Renombrar las columnas para mayor claridad
faltantes_regular.columns = ['Columna', 'Cantidad Faltantes']
faltantes_playoff.columns = ['Columna', 'Cantidad Faltantes']

# Imprimir los resultados de los valores faltantes
print("Resultados de valores faltantes en la Temporada Regular:")
print(faltantes_regular)

print("\nResultados de valores faltantes en los Playoffs:")
print(faltantes_playoff)

# Paso 5: Crear copias de los DataFrames originales para no sobrescribirlos
df_regular_faltantes = df_regular.copy()
df_playoff_faltantes = df_playoff.copy()

# Rellenar valores faltantes
if not faltantes_regular.empty:
    df_regular_faltantes.fillna('Faltante', inplace=True)
else:
    print("No hay valores faltantes en la Temporada Regular.")

if not faltantes_playoff.empty:
    df_playoff_faltantes.fillna('Faltante', inplace=True)
else:
    print("No hay valores faltantes en los Playoffs.")

# Conteo de valores modificados en cada dataset
conteo_modificados_regular = (df_regular_faltantes == 'Faltante').sum().sum()
conteo_modificados_playoff = (df_playoff_faltantes == 'Faltante').sum().sum()

# Imprimir el conteo de los valores modificados
print(f"Conteo de valores modificados en la Temporada Regular: {conteo_modificados_regular}")
print(f"Conteo de valores modificados en los Playoffs: {conteo_modificados_playoff}")

# Paso 6: Exportar los datasets modificados
ruta_regular_modificado = '/content/drive/My Drive/Colab Notebooks/Examen 01/df_regular_modificado.csv'
ruta_playoff_modificado = '/content/drive/My Drive/Colab Notebooks/Examen 01/df_playoff_modificado.csv'

df_regular_faltantes.to_csv(ruta_regular_modificado, sep=';', index=False)  # Asegúrate de incluir el separador
df_playoff_faltantes.to_csv(ruta_playoff_modificado, sep=';', index=False)  # Asegúrate de incluir el separador

# Confirmación de exportación
print(f"Dataset modificado de la Temporada Regular exportado a: {ruta_regular_modificado}")
print(f"Dataset modificado de los Playoffs exportado a: {ruta_playoff_modificado}")

### CITAS ###
# 1. Pandas Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
# 2. Google Colab - Connecting to Google Drive: https://colab.research.google.com/notebooks/io.ipynb
# 3. Pandas read_csv documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
# 4. Pandas Documentation on DataFrame.to_csv: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html

```
